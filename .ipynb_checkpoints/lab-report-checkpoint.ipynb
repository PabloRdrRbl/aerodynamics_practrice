```python
# Importing the different libraries needed
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
csv_path = 'naca_0018.csv'
```

```python
csv_labels = np.genfromtxt(csv_path, delimiter=',', usecols=0, dtype=str)
csv_data = np.genfromtxt(csv_path, delimiter=',', dtype=float)[:,1:]
```

```python
csv_data.shape # Ensure you've got the right shape for the narray
```

```python
def plot_voltages(x_data, upper_data, lower_data, aoa):
    fig, ax = plt.subplots(figsize=(6, 4), dpi=800)

    ax.plot(x_data, upper_data, 'r.--', label='upper')
    ax.plot(x_data, lower_data, 'b.--', label='lower')

    ax.set_xlim(0, 1)
    ax.set_ylim(-4, 2)

    #ax.fill_between(x_data, upper_data, lower_data, facecolor='green', alpha=0.25, label='lift')

    ax.set_title("Voltages with AOA$ = {0}^\circ$".format(aoa))
    ax.legend(loc='lower right')
    ax.set_xlabel('x position in the airfoil')
    ax.set_ylabel('voltage meassured')

    ax.grid()
    
    return fig
```

```python
plot_voltages(csv_data[0], csv_data[1], csv_data[5], 0)
plot_voltages(csv_data[0], csv_data[2], csv_data[6], 7)
plot_voltages(csv_data[0], csv_data[3], csv_data[7], 14)
plot_voltages(csv_data[0], csv_data[4], csv_data[8], 21)
```
