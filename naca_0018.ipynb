## Meassuring pressure on NACA 0018 airfoil

The following notebook shows how to write from CSV file into a numpy array. Then we plot some of de voltage meassurements taken in the wind tunnel.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from io import StringIO
```

```python
csv_path = '/Users/Pablo/Desktop/naca_0018.csv'
```

```python
csv_labels = np.genfromtxt(csv_path, delimiter=',', usecols=0, dtype=str)
csv_data = np.genfromtxt(csv_path, delimiter=',', dtype=float)[:,1:]
```

```python
csv_data.shape ## Ensure you've got the right shape for the narray
```

```python
fig_1, ax1 = plt.subplots(figsize=(6, 4), dpi=800)

ax1.plot(csv_data[0], csv_data[2], 'b.--', label='upper')
ax1.plot(csv_data[0], csv_data[6], 'r.--', label='lower')

ax1.set_xlim(0, 1)
ax1.set_ylim(0, 14)

ax1.fill_between(csv_data[0], csv_data[2], csv_data[6], facecolor='green', alpha=0.25, label='lift')

ax1.set_title('Voltages with AOA$ = 7^\circ$')
ax1.legend(loc='lower right')
ax1.set_xlabel('x position in the airfoil')
ax1.set_ylabel('voltage meassured')

ax1.grid()
```
